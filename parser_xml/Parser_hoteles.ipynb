```python
import xml.etree.ElementTree as ET
import csv
import pandas as pd
import re
import html
```

```python
tree = ET.parse("C:\\Users\\cx02386\\OneDrive - Telefonica\\Master\\TFM\\ficheros\\alojamientos_v1_es.xml")
root = tree.getroot()
```

```python
# Dict pandas: {0 : ["dfvbh", ....], ...}
dict_parsed_for_pd = {}
counter = 0
for service in root.findall('service'):
        
    categoria = ''
    ls = list(service.find('extradata').find('categorias'))
    #Comprobar hoteles donde la categoria este vacia
    if len(ls) == 0:
        categoria = "Sin Categoria"
    else:
        hijo = ls[0]
        item = hijo.find('item[2]')
        categoria = item.text
        
         
    l = [         
        html.unescape (service.find('basicData').find('name').text), 
        service.find('basicData').find('email').text,
        html.unescape (service.find('basicData').find('phone').text),
        service.find('basicData').find('body').text,
        html.unescape (service.find('basicData').find('web').text),
        html.unescape (service.find('geoData').find('latitude').text),
        html.unescape (service.find('geoData').find('longitude').text),
        html.unescape (categoria) 
    ]

       
    dict_parsed_for_pd[counter] = l
    counter += 1
```

```python
df = pd.DataFrame.from_dict(dict_parsed_for_pd, orient='index', columns=['name', 'email', 'phone', 'body', 'web', 'latitude', 'longitude','categorias'])
```

```python
cleanr = re.compile('<.*?>')
df["newBody"] = html.unescape (df["body"].apply(lambda x: re.sub(cleanr, '', str(x))))
del(df["body"])
df.head(20)
```
